```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Data Preparation
# Membaca dataset
file_path = "../data iris.csv"
df = pd.read_csv(file_path, delimiter=";")

# Memilih variabel X dan y
X = df[['sepal_length']].values  # Variabel independen
y = df['petal_length'].values    # Variabel dependen

# Menambahkan bias (intercept)
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Menambahkan kolom 1 di depan

# 2. Model Training dengan Normal Equation
# Menghitung koefisien regresi
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# 3. Prediksi
# Menggunakan model regresi untuk memprediksi nilai petal_length
y_pred = X_b.dot(theta)

# 4. Visualisasi
plt.scatter(X, y, color='blue', label='Data Asli')  # Scatter plot data asli
plt.plot(X, y_pred, color='red', label='Regresi Linear')  # Garis regresi
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Linear Regression: Sepal Length vs Petal Length")
plt.legend()
plt.show()

# Menampilkan persamaan regresi
a, b = theta
print(f"Persamaan regresi: y = {a:.3f} + {b:.3f} * X")
tes = np.corrcoef(X,y, rowvar=False)
print(tes)
```
