```python
# Fungsi untuk membersihkan teks
def clean_text(text):
    text = text.lower()  # Ubah ke huruf kecil
    text = re.sub(r'\d+', '', text)  # Hapus angka
    text = text.translate(str.maketrans('', '', string.punctuation))  # Hapus tanda baca
    text = text.strip()  # Hapus spasi di awal/akhir
    return text
```

```python
# Terapkan fungsi cleaning
data["cleaned_content"] = data["content"].apply(clean_text)
# Lihat hasilnya
data[["content", "cleaned_content"]].head()
```

```python
stop_words = set(stopwords.words("indonesian"))  # Stopwords bahasa Indonesia
```

```python
# Fungsi untuk menghapus stopwords
def remove_stopwords(text):
    words = text.split()
    filtered_text = [word for word in words if word not in stop_words]
    return " ".join(filtered_text)
```

```python
# Terapkan fungsi stopwords
data["cleaned_content"] = data["cleaned_content"].apply(remove_stopwords)
```
