```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fungsi untuk memuat data
def load_data(filename):
    data = pd.read_csv(filename, delimiter=';')
    features = data.iloc[:, :-1]
    labels = data.iloc[:, -1]
    return features, labels

# Load data
filename = 'data iris.csv'
features, labels = load_data(filename)
```

```python
# Scatter Plot (sepal length vs sepal width)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=features.iloc[:, 0], y=features.iloc[:, 1], hue=labels, palette="Set1")
plt.title("Scatter Plot - Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend(title="Classes")
plt.show()
```

```python
# Boxplot untuk setiap fitur
plt.figure(figsize=(10, 6))
features_melted = features.melt(var_name="Features", value_name="Values")
sns.boxplot(x="Features", y="Values", data=features_melted)
plt.title("Box Plot of Features")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
features.hist(bins=15, color="skyblue", edgecolor="black", grid=False, figsize=(10, 6))
plt.suptitle("Histogram of Features", y=1.02, fontsize=16)
plt.show()
```

```python
# Pie Chart untuk distribusi label
plt.figure(figsize=(8, 8))
label_counts = labels.value_counts()
plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Class Distribution")
plt.show()
```
